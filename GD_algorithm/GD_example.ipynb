GD is used to estimate parameter, we use GD to descent to lowest point in a loss function.

### Example: Linear function: y=kx+b




GD can fit a line to data by finding the optimal values for the intercept and slope(two parameters)

Let's start with intercept, and assume slope is fixed. Here, GD is very efficient to do only a few calculations far from the optimal solution. GD identifies the optimal value by taking big steps when it is far away, and tiny step steps when it is increasingly close to optimal value. It’s an efficient way to save computational resources. Whether the big step or the tiny step, the size of the step should be realated to the derivative, incase we waste too much computational resources or miss the optimal point. For demonstration, we assume there is only one parameter intercept.

First, we randomly pick a value for intercept, it’s an initial guess that gives Gradient Descent something to start.

Second, we calculate the sum of the square residuals(SSE), to evaluate how well the line fits the data. In ML, SSE is a type of loss function(the difference between the predicted value and the observed value)

#### Note: sum of the square residuals=residual sum of squares=SSE

$y_i=$ observed value

$\widehat{y}_i=$predicted value

$\overline{y}_i=$average observed value

$$SST=\sum_{i=1}^n\left(y_i-\overline{y}\right)^2$$

$$SSR=\sum_{i=1}^n\left(\widehat{y}_i-\overline{y}\right)^2$$

$$SSE=\sum_{i=1}^n\left(y_i-\widehat{y}_i\right)^2=\sum_{i=1}^ne_i^2$$

$$SST=SSR+SSE=\sum_{i=1}^n\left(y_i-\overline{y}\right)^2=\sum_{i=1}^n\left(\widehat{y}_i-\overline{y}\right)^2+\sum_{i=1}^n\left(y_i-\widehat{y}_i\right)^2$$


##### SST = total sum of squares
• total variabilitv of  $Y$

• depends on the response $Y$ only, not on the form of the model

##### SSR = regression sum of squares
• variability of $Y$ explained by $X_1, ..., X_n$

##### SSE = error (residual) sum of squares
• variability of $Y$ not explained by the $X$


We can write an equation of SSE function and take derivative at any value of the intercept.

GD will use the derivative of SSE function to find where the SSE is lowers. GD determines the step size by multiplying the slope by a small number called the **learning rate(0.1)**

After claculation with the initial value of intercept, how to find the next intercept? In GD, **step size= derivative* learning rate**, so **new intercept = old intercept – derivative*learning rate.**

First, we randomly pick a value for intercept, it’s an initial guess that gives Gradient Descent something to start.

assume k is a constant. 

Predicted price: 

$\widehat{y}_1=k*(s_1)+b=ks_1+b$

$\widehat{y}_2=k*(s_2)+b=ks_2+b$

$\widehat{y}_3=k*(s_3)+b=ks_3+b$

$\widehat{y}_4=k*(s_3)+b=ks_4+b$

$\widehat{y}_5=k*(s_3)+b=ks_5+b$

$SSE=\sum_{i=1}^5\left(y_i-\widehat{y}_i\right)^2=f(b)$


$$\dfrac{d}{d(parameter)}SSE=\dfrac{d}{d(intercept)}SSE=\dfrac{d}{db}f(b)=\dfrac{d}{dx}f(x)=f'(x)$$

initialize intercept: $b_1=0$

$$b_1=0 \Rightarrow f'(x)|_{x=b_1}$$

step size=derivative* learning rate $\Rightarrow$ new intercept = old intercept – learning rate*derivative

$$b_2=b_1-rate*f'(x)|_{x=b_1}$$

$$b_2 \Rightarrow f'(x)|_{x=b_2}$$

$$b_3=b_2-rate*f'(x)|_{x=b_2}$$

$$b_3 \Rightarrow f'(x)|_{x=b_3}$$

GD will stop when the step size is very close to 0, because **step size= derivative* learning rate**, when the derivative is close to 0, the step size will close to 0. In practice, the step size smaller than **0.001** is sufficient. And normally, **maximum number of steps = 1000** or greater. So GD will stop when it meets the thresholds of step size or number of steps. (要么步长小到一定值，要么步数到达一定值） 

The old intercept starts from the initial value of the parameter, and it doesn’t matter whether the initial value is smaller or bigger than the optimal value. The equation ensures that we approach the optimal value from two sides. Imaging the SSE function is a function of intercept, it’s a U shape function, if the derivative is negative, the value of intercept is on the left side of the optimal value, if it’s positive, the value is on the right side of the optimal value. 

GD can be very sensitive to learning rate. In practive, the learning rate determined automatically by starting large and getting smaller with each step. Learning rate is not an invariant.

If the slope is also a variant, we can take partial derivative at any value of slope. If there are many parameters in the loss function, that’s ok, we can take partial derivative with respect to those parameters with chain rule one by one till we find the best parameter candidates.
